```python
import requests

# Делаем GET запрос к публичному API
response = requests.get('https://api.github.com/users/slk-oss')

# Смотрим статус
print(response.status_code)

# Смотрим данные в JSON
data = response.json()
print(data)
```

```python
import json

print(json.dumps(data, indent=2))
```

```python
# Достаём конкретные поля
print("Login:", data['login'])
print("Public repos:", data['public_repos'])
print("Followers:", data['followers'])
print("Location:", data['location'])
print("Created at:", data['created_at'])
```

```python
# Получаем список репозиториев
repos_response = requests.get('https://api.github.com/users/slk-oss/repos')
repos = repos_response.json()

# Смотрим сколько репозиториев
print(f"Количество репозиториев: {len(repos)}")

# Выводим название и язык каждого
for repo in repos:
    print(f"{repo['name']} — {repo['language']} — ⭐ {repo['stargazers_count']}")
```

```python
import pandas as pd

# Превращаем JSON в DataFrame
df_repos = pd.DataFrame(repos)

# Выбираем нужные столбцы
df_repos = df_repos[['name', 'language', 'stargazers_count', 
                       'forks_count', 'created_at', 'updated_at']]

print(df_repos)
```

```python
import plotly.express as px

df_repos.to_csv('github_repos.csv', index=False)

# График — когда создавались репозитории
df_repos['created_at'] = pd.to_datetime(df_repos['created_at'])
df_repos['name_short'] = df_repos['name'].str[:15]

fig = px.bar(df_repos, x='name_short', y='stargazers_count',
             color='language', title='My GitHub repositories')
fig.show()
```
