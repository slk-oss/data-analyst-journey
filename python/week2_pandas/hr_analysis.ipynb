```python
import pandas as pd
import numpy as np

data = {
    "emp_id": [101,102,103,104,105,106,107,108,109,110],
    "name": [
        "Иванов А.","Петрова Б.","Сидоров В.","Козлова Г.","Морозов Д.",
        "Волкова Е.","Новиков Ж.","Федорова З.","Алексеев И.","Дмитриева К."
    ],
    "department": [
        "dev","dev","analytics","dev","marketing",
        "analytics","marketing","dev","analytics","dev"
    ],
    "position": [
        "junior","middle","senior","senior","middle",
        "junior","senior","junior","middle","middle"
    ],
    "salary": [
        80000,150000,200000,220000,120000,
        70000,180000,85000,160000,140000
    ],
    "hire_date": [
        "2023-01-15","2021-06-01","2019-03-10","2018-11-20","2022-02-28",
        "2023-09-01","2020-07-15","2023-04-10","2021-01-20","2022-08-05"
    ],
    "city": [
        "Москва","Москва","Санкт-Петербург","Москва","Казань",
        "Санкт-Петербург","Москва","Казань","Москва","Санкт-Петербург"
    ],
    "rating": [
        4.2,3.8,4.9,4.5,3.1,
        4.7,2.9,4.0,3.5,4.3
    ]
}

df = pd.DataFrame(data)
```

# 4 Расчеты и выводы

## Блок A

```python
# 1. Покажи всех сотрудников из dev с зарплатой выше 100 000
print("\nСотрудники из dev с зарплатой выше 100 000:")
result = df[(df['department'] == 'dev') & (df['salary'] > 100000)]
display(result)

# 2.  Найди сотрудников с рейтингом ниже 3.5 — это кандидаты на performance review
print("\nКандидаты на performance review (рейтинг ниже 3.5):")
candidates = df[df['rating'] < 3.5]
display(candidates)

# 3.  Отсортируй всех по зарплате (убывание), а при равной зарплате — по рейтингу (убывание)
print("\nОтсортированные по зарплате (убывание):")
display(df.sort_values(by=['salary', 'rating'], ascending=[False,False]))
```

## Блок B

```python
# 4. Средняя зарплата по каждому отделу
print(f"\nСредняя зарплата по каждому отделу: {df.groupby('department')['salary'].mean()}")

# 5. Количество сотрудников по городам
print(f"\nКоличество сотрудников по городам: {df['city'].value_counts()}")

# 6. Для каждого отдела покажи: минимальную, максимальную и среднюю зарплату одним вызовом
print(f"\nминимальная, максимальная и средняя зарплата по отделу: \n{df.groupby('department')['salary'].agg(['min','max','mean'])}")
```

```python
# 7. Добавь колонку experience_years — сколько полных лет сотрудник работает (от hire_date до сегодня)
df['hire_date'] = pd.to_datetime(df['hire_date'])
df['experience_years'] = (pd.Timestamp.now() - df['hire_date']).dt.days//365
print(f"\nсколько полных лет работает каждый сотрудник: \n{df[['name', 'experience_years']]}")
# 8. Найди самого высокооплачиваемого сотрудника в каждом отделе

print("\nсамый высокооплачиваемый сотрудник в каждом отделе")
top_indices = df.groupby('department')['salary'].idxmax()
top_by_idxmax = df.loc[top_indices]
display(top_by_idxmax)

# 9. Покажи сотрудников, чья зарплата выше средней по их отделу
print("сотрудники, чья зарплата выше средней по их отделу")
avg_salary_by_dept = df.groupby('department')['salary'].transform('mean')
display(df[df['salary'] > avg_salary_by_dept])
```

# Анализ

```python
junior_high_earners = df[(df['position'] == 'junior') & (df['salary'] > avg_salary_by_dept)]
print("\njunior сотрудники с зарплатой выше средней по отделу:")
print(junior_high_earners.shape[0]) # таких нет
display(junior_high_earners[['name', 'department', 'position', 'salary']])

middle_high_earners = df[(df['position'] == 'middle') & (df['salary'] > avg_salary_by_dept)]
print("\nmiddle сотрудники с зарплатой выше средней по отделу:")
print(middle_high_earners.shape[0]) 
display(middle_high_earners[['name', 'department', 'position', 'salary']])

senior_high_earners = df[(df['position'] == 'senior') & (df['salary'] > avg_salary_by_dept)]
print("\nsenior сотрудники с зарплатой выше средней по отделу:")  
print(senior_high_earners.shape[0])
display(senior_high_earners[['name', 'department', 'position', 'salary']])
```
